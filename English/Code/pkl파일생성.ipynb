```python
import json
import pickle

# JSON 파일 로드
json_path = "../데이터/highschool_english_dataset.json"
with open(json_path, "r", encoding="utf-8") as f:
    dataset = json.load(f)

# 질문과 정답을 리스트로 변환
faiss_data = []
for _, content in dataset.items():
    questions = content.get("questions", {}).get("cleaned_text", "").split("\n")
    answers = content.get("answers", {})

    for i, question in enumerate(questions, 1):
        if str(i) in answers:  # 정답이 존재하는 경우
            faiss_data.append({"question": question.strip(), "answer": answers[str(i)]})

# `faiss_data.pkl` 저장
pkl_path = "faiss_data.pkl"
with open(pkl_path, "wb") as f:
    pickle.dump(faiss_data, f)

print(f"✅ {pkl_path} 생성 완료! 총 {len(faiss_data)}개의 데이터가 저장됨.")
```

```python
import json
import pickle
import re

# ✅ JSON 파일 경로
json_path = "../데이터/highschool_english_dataset.json"

# ✅ JSON 파일 로드
with open(json_path, "r", encoding="utf-8") as f:
    json_data = json.load(f)

# ✅ 모든 문제 데이터를 리스트로 변환
faiss_data = []

# 🔍 JSON의 각 파일(시험지) 순회
for key, content in json_data.items():
    if isinstance(content, dict) and "questions" in content and "answers" in content:
        raw_text = content["questions"].get("raw_text", "")
        answers = content["answers"]
        
        # ✅ 문제 번호 + 선택지 추출 (정규 표현식 사용)
        question_pattern = r"(\d+)\.\s(.+?)(?=\s\d+\.\s|$)"  # 문제 번호를 기준으로 분리
        matches = re.findall(question_pattern, raw_text, re.DOTALL)

        for q_number, q_text in matches:
            q_number = q_number.strip()  # 문제 번호 정리
            q_text = q_text.strip()  # 문제 내용 정리
            
            # ✅ 선택지 추출 (① ~ ④ 형태)
            option_pattern = r"① (.*?) ② (.*?) ③ (.*?) ④ (.*?)$"
            option_match = re.search(option_pattern, q_text, re.DOTALL)

            if option_match:
                options = list(option_match.groups())  # 선택지 리스트로 변환
                q_text = re.sub(option_pattern, "", q_text).strip()  # 선택지 부분 제거한 문제 텍스트
            else:
                options = []  # 선택지가 없을 경우 빈 리스트

            # ✅ 정답 매핑
            correct_answer = answers.get(q_number, None)
            if correct_answer:
                try:
                    correct_answer = int(correct_answer) - 1  # 1~4 -> 0~3 인덱스로 변환
                    correct_answer = options[correct_answer] if 0 <= correct_answer < len(options) else None
                except ValueError:
                    correct_answer = None

            # ✅ 데이터 저장
            if options and correct_answer:
                faiss_data.append({
                    "question": q_text,
                    "options": options,
                    "answer": correct_answer
                })

# ✅ `faiss_data.pkl` 저장
faiss_data_path = "faiss_data.pkl"
with open(faiss_data_path, "wb") as f:
    pickle.dump(faiss_data, f)

print(f"✅ `faiss_data.pkl` 생성 완료! 총 문제 개수: {len(faiss_data)}")
```
